```python
import json
with open(file_path, 'r') as file:
    data = json.load(file)


df = pd.json_normalize(data, sep='_')

def is_valid_indian_phone_number(phone_number):
    phone_number = str(phone_number).strip()
    
    if phone_number.startswith('+91'):
        phone_number = phone_number[3:]
    elif phone_number.startswith('91'):
        phone_number = phone_number[2:]
    
    return len(phone_number) == 10 and phone_number.isdigit() and phone_number[0] in '6789' and 6000000000 <= int(phone_number) <= 9999999999


df['isValidMobile'] = df['phoneNumber'].apply(is_valid_indian_phone_number)


valid_phone_count = df['isValidMobile'].sum()


print(f"Number of valid phone numbers found: {valid_phone_count}")

print(df[['phoneNumber', 'isValidMobile']])
```

```python
columns_of_interest = ['patientDetails_firstName', 'patientDetails_lastName', 'patientDetails_birthDate']

missing_percentages = {}

for column in columns_of_interest:
    total_values = len(df)
    missing_values = df[column].isnull().sum() + (df[column] == '').sum()
    missing_percentage = (missing_values / total_values) * 100
    missing_percentages[column] = round(missing_percentage, 2)

for column, percentage in missing_percentages.items():
    print(f"Missing percentage for {column}: {percentage}%")
```
